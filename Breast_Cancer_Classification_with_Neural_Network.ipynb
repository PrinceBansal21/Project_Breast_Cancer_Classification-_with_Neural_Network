```python
# importing required libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
```

```python
# loading the data from sklearn

breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
```

```python
print(breast_cancer_dataset)
```

```python
#loading data to data frame

data_frame = pd.DataFrame(breast_cancer_dataset.data, columns = breast_cancer_dataset.feature_names)
```

```python
data_frame.head()
```

```python
# adding the target column to the data frame

data_frame['label'] = breast_cancer_dataset.target
```

```python
data_frame.tail()
```

```python
# getting some information about the data

data_frame.info()
```

```python
data_frame.isnull().sum()
```

```python
# statistical measures about the data

data_frame.describe()
```

```python
# checking the distribution of Target variable

data_frame['label'].value_counts()
```

1 -> Benign (Can be treated easily)
0 -> Malignant (Difficult to treat)

```python
data_frame.groupby('label').mean()
```

```python
# separating the features and target

X = data_frame.drop(columns = 'label', axis=1)
Y = data_frame['label']
```

```python
print(X)
print(Y)
```

```python
# splitting the data into training data & testing data

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
# Standardize the data

from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
X_train_std = scaler.fit_transform(X_test)
X_test_std = scaler.transform(X_test)
```

```python
# Building the neural network
# importing the required libraties for creating neural network

import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
```

```python
# setting up the layers for Neural Network

model = keras.Sequential([
                          keras.layers.Flatten(input_shape = (30,)),
                          keras.layers.Dense(20, activation = 'relu'),
                          keras.layers.Dense(2, activation = 'sigmoid')
])
```

```python
# compling the Neural Network

model.compile(optimizer = 'adam',
              loss = 'sparse_categorical_crossentropy',
              metrics = ['accuracy'])
```

```python
# training the Neural Network

history = model.fit(X_train_std, Y_train, validation_split=0.1, epochs = 20)
```

```python
# Visualize accuracy and loss

plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title('model_accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')

plt.legend(['training data', 'validation data'], loc = 'lower right')
```

```python

plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.title('model_loss')
plt.ylabel('loss')
plt.xlabel('epoch')

plt.legend(['training data', 'validation data'], loc = 'lower right')
```

```python
# Accuracy of model on test data

loss, accuracy = model.evaluate(X_test_std, Y_test)
print(accuracy)
```

```python
Y_pred = model.predict(X_test_std)
```

model.predict() gives the prediction probability of each data for that data 

```python
# argmax function, just an example for next code to return the max value index position

my_list = [0.25, 0.56]

index_of_max_value = np.argmax(my_list)
print(my_list)
print(index_of_max_value)
```

```python
# converting the prediction probability to class labels

Y_pred_labels = [np.argmax(i) for i in Y_pred]
print(Y_pred_labels)
```

```python
# Buidling the predictive system

input_data = (0.57, 17.77, 132.9, 1326, 0.08474, 0.07864, 0.0869, 0.07017, 0.1812, 0.05667, 0.5434, 0.7339, 3.398, 74.08, 0.005225, 0.01308, 0.01786, 0.0134, 0.01389, 0.003532, 24.99, 23.41, 158.8, 1956, 0.1238, 0.1866, 0.2416, 0.186, 0.275, 0.08902)

# change the input data to numpy array
input_data_as_numpy_array = np.asarray(input_data)

#reshape the numpy array as we are predicting for one data point
input_data_reshaped = input_data_as_numpy_array.reshape(1, -1)

# standardizing the input data
input_data_std = scaler.transform(input_data_reshaped)

prediction = model.predict(input_data_std)
print(prediction)

prediction_label = [np.argmax(prediction)]
print(prediction_label)

if(prediction_label[0] == 0):
  print('The tumor is malignant')

else:
  print('The tumor is Benign')
```
